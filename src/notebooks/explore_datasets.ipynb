```r
PRE_DIR  <- "../../data/preprocessed_rds_files"
CLIN_DIR <- "../../data/raw_clinical_rds_files"
DEG_DIR     <- "../../data/deg_rds_files"
RESULTS_DIR <- "../../results/deg"
DATA_DIR    <- "../../data"

datasets        <- c("GSE42568", "GSE21653", "GSE20711", "GSE88770")
SIGNATURE_GENES <- c("TSLP", "BIRC5", "S100B", "MDK", "S100P", "RARRES3", "BLNK", "ACO1")

RAW_RFS_COLS <- c(
    GSE42568 = "relapse free survival time_days:ch1",
    GSE21653 = "dfs time (months):ch1",
    GSE20711 = "t.rfs:ch1",
    GSE88770 = "drfs_or_last_contact_years:ch1"
)
```

```r
view_preprocessed <- function(name) {
    expr     <- readRDS(file.path(PRE_DIR, paste0(name, "_expression_matrix.rds")))
    clinical <- readRDS(file.path(PRE_DIR, paste0(name, "_clinical_metadata.rds")))
    clinical
}

view_raw_clinical <- function(name) {
    df <- readRDS(file.path(CLIN_DIR, paste0(name, "_raw_clinical.rds")))
    cat("Dataset:", name, "|", nrow(df), "samples x", ncol(df), "columns\n\n")
    for (col in colnames(df)) {
        vals     <- df[[col]]
        n_unique <- length(unique(vals))
        cat(sprintf("  %-50s [%d unique]  ", col, n_unique))
        if (n_unique <= 15) {
            cat(paste(sort(unique(vals)), collapse = " | "))
        } else {
            cat(class(vals), "— e.g.", paste(head(unique(vals), 3), collapse = ", "), "...")
        }
        cat("\n")
    }
    invisible(df)
}

summarise_rfs_time <- function(name) {
    df   <- readRDS(file.path(CLIN_DIR, paste0(name, "_raw_clinical.rds")))
    col  <- RAW_RFS_COLS[[name]]
    vals <- as.numeric(df[[col]])
    cat("RFS time raw distribution —", name, "\n")
    cat("Column:", col, "\n\n")
    print(summary(vals))
    invisible(vals)
}
```
