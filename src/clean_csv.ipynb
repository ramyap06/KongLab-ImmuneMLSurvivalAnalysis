```python
import pandas as pd
```

```python
expr_filename = "../datasets/csv_files/deg_expression_matrix_train.csv"
clinical_filename = "../datasets/csv_files/clinical_metadata_train.csv"

# load datasets
expr = pd.read_csv(expr_filename)
clinical = pd.read_csv(clinical_filename)

expr.tail()
```

```python
# rename column for clarity
#expr_filtered = expr.iloc[:, 1:]
clinical_filtered = clinical.drop(columns=['sample_id'])
clinical_filtered.rename(columns={'Unnamed: 0': 'sample_name'}, inplace=True)
expr_filtered = expr.rename(columns={'Unnamed: 0': 'sample_name'})
```

```python
# save datasets
expr_filtered.to_csv(expr_filename, index=False)
clinical_filtered.to_csv(clinical_filename, index=False)
```
