```python
from upsetplot import from_contents, plot
import matplotlib.pyplot as plt
```

```python
paper_genes = ['RARRES3', 'ACO1','TSLP', 'BIRC5', 'S100B', 'MDK', 'S100P', 'BLNK', ]
cox_genes = ['UBE2S', 'S100P']
rsf_genes = ['SOX4', 'RARRES3', 'ACO1', 'HBB', 'CFB', 'FBXL16', 'CYSTM1', 'ARF1']
gbs_genes = ['CDC20', 'PLK2', 'PYGL', 'EVL', 'SOX4', 'FBXL5', 'MICB', 'JUP']
ssvm_genes = ['S100P', 'CXCL9', 'VTCN1', 'GRP', 'S100A14', 'S100A7', 'SLC39A6', 'ZBTB16']
```

```python
data = {
    'Paper 8-Gene Signature': paper_genes,
    'Cox Regression': cox_genes,
    'Random Survival Forest': rsf_genes,
    'Gradient Boosting Survival': gbs_genes,
    'Survivial SVM': ssvm_genes
}
upset_data = from_contents(data)
```

```python
plot(
    upset_data,
    facecolor="#4C72B0",
    show_counts=False,
    sort_by="cardinality"
)
plt.title("Gene Signature Overlap Across Models", fontsize=14)
plt.savefig("../visuals/sig_genes_overlap.png", dpi=150, bbox_inches="tight")
```
