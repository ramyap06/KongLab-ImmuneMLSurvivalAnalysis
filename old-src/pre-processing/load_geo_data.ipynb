```r
library(affy)
library(GEOquery)
library(tidyverse)
```

```r
# defining datapath
CEL_DIR <- "../../data/cel_files"
TAR_DIR <- "../../data/tar_files"

# defining datasets
geo_files <- c("GSE42568", "GSE21653", "GSE20711", "GSE88770")
```

```r
load_data <- function(name) {
    # get supplementary files
    dir.create("../data/tar_files", recursive = TRUE, showWarnings = FALSE)
    getGEOSuppFiles(name, baseDir="../data/tar_files")

    # untar file
    dir.create("../data/cel_files", recursive = TRUE, showWarnings = FALSE)
    untar(paste0("../data/tar_files/", name, "/", name, "_RAW.tar"), exdir = paste0("../data/cel_files/", name))
}
```

```r
# training
load_data("GSE42568")

# validation
load_data("GSE21653")
load_data("GSE20711")
load_data("GSE88770")
```
